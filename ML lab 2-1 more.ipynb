```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```python
tf.set_random_seed(777) # for reproducibility
```

# 1. X and Y data 미리 설정하는 경우

## 1) build graph using TF operation

```python
x_train = [1,2,3]
y_train = [1,2,3]
```

### "y = W*x + b" 만족시키는 W b 값 찾기

- tf.Variable : trainable 한 valuable
- tf.random_normal([1]) : shape=1인 random_normal

```python
W = tf.Variable(tf.random_normal([1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')
```

### 우리가 찾는 모델 => y = W*x + b

![title](reduce_mean.jpg)

```python
hypothesis = x_train * W + b

# cost/loss function
cost = tf.reduce_mean(tf.square(hypothesis - y_train))
```

### minimize

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)
```

### prepare session

```python
# launch the graph in session
sess = tf.Session()

# Inintializes global variables in the graph
sess.run(tf.global_variables_initializer())
```

### fit the line

<img src='sess.run(train).jpg' alt='drawing' width='200'/>

- sess.run(train) 하면 노드들이 모두 이어져 결국 W와 b 값을 알아옴

```python
for step in range(2001): #2000번 반복 실행
    sess.run(train) # train 실행
    if step % 20 == 0 : # 20번째 배수 실행마다 print
        print(step, sess.run(cost), sess.run(W), sess.run(b))
        # 몇번째, cost값(점점 줄어야함), W값(1에 가깝고), b값(0에 가까워야함) 
```

- 여기까지 미리 x_trian과 y_train값을 설정해 사용하는 방법
- 이제 placeholder로 x,y 값 설정하고 feed_dict를 통해 나중에 x,y 값 결정하는 방법

# 2. X, Y 값 후에 설정하기 (placeholder, feed_dict)

### "y = W*x + b" 만족시키는 W b 값 찾기

```python
W = tf.Variable(tf.random_normal([1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')
```

### X와 Y data

- shape=[None] : None = n이라고 생각, None이라고 해두면 내가 숫자 넣고 싶은대로 1개,2개, ... ,n개 막 넣어도됨

```python
# 미리 값 넣지 않고 placeholder로 설정
# 후에 feed_dict로 값 입력
X = tf.placeholder(tf.float32, shape=[None])
Y = tf.placeholder(tf.float32, shape=[None])
```

### our model : Y = W*x + b

```python
hypothesis = X*W + b

# cost/loss function
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

### Minimize

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)
```

### Prepare session

```python
# launch the graph in a session
sess = tf.Session()

#Initializes global variables in the graph
sess.run(tf.global_variables_initializer())
```

### fit the line

```python
for step in range(2001):
    cost_val, W_val, b_val, _ = \
        sess.run([cost, W, b, train], feed_dict={X:[1,2,3], Y: [1,2,3]})
    if step%20 ==0:
        print(step, cost_val, W_val, b_val)
```

### testing our model

```python
print(sess.run(hypothesis, feed_dict={X: [5]}))
print(sess.run(hypothesis, feed_dict={X: [2.5]}))
print(sess.run(hypothesis, feed_dict={X: [1.5, 3.5]}))
```

## Fit the line with new training data

```python
for step in range(2001):
    cost_val, W_val, b_val, _ = \
        sess.run([cost, W, b, train], feed_dict={X: [1,2,3,4,5], Y:[2.1,3.1,4.1,5.1,6.1]})
    if step%20 == 0:
        print(step, cost_val, W_val, b_val)
```

### training our model

```python
print(sess.run(hypothesis, feed_dict={X: [5]}))
print(sess.run(hypothesis, feed_dict={X: [2.5]}))
print(sess.run(hypothesis, feed_dict={X: [1.5, 3.5]}))
```
